```python
# convert into vectors
from sentence_transformers import SentenceTransformer
```

```python
encoder = SentenceTransformer("all-mpnet-base-v2")
vectors = encoder.encode(df.text)
vectors.shape
```

```python
dim = vectors.shape[1]
dim
```

```python
import faiss
index = faiss.IndexFlatL2(dim)
index
```

```python
index.add(vectors)
```

```python
search_query = "I want to buy something"
vec = encoder.encode(search_query)
vec.shape
```

```python
import numpy as np
svec = np.array(vec).reshape(1,-1)
svec.shape
```

```python
distances, I = index.search(svec,k=2)
```

## Stuff method and map-reduce method

- `Stuff method is the most simplest method of all where after giving the embeddings to the vector database, we just find out the similarity based on the search query and we get some chunks of data lets say we got 2 chunks based on the similarity of our search query. And then we pass those chunks to the open Ai api key which do has some limit like upto 1000 tokens. So when we are passing the chunks, we need to be careful that our chunks should not exceed the limit of openai API.`

- `Map-Reduce method is something where we pass the chunks to individual LLM. We may do some more calls in this method when compared to stuff mehtod. we just do one call in stuff method`
